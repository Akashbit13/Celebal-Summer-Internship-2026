## Objective

The objective of this project is to analyze the Sample Superstore sales dataset using SQL. The analysis includes exploring the dataset, applying filters, performing aggregations, solving business-related queries, and validating data quality to gain useful business insights.

## Step 1: Create Database

CREATE DATABASE superstore_db;

USE superstore_db;

---> Create Table

The table was created automatically using the **MySQL Table Data Import Wizard** while importing the dataset.

---> Load Dataset

The **SampleSuperstore.csv** dataset was successfully imported into the **superstore** table using the MySQL Table Data Import Wizard.

## Step 3:Apply WHERE filters (region, category, date, sales)

-- Region Filter--

SELECT *
FROM superstore
WHERE region='West';
### Output


-- Category Filter--

SELECT *
FROM superstore
WHERE category='Furniture';
### Output


-- Sales Greater Than 1000--

SELECT *
FROM superstore
WHERE sales>1000;
### Output


-- Profit Less Than 0 (Loss Orders)--

SELECT *
FROM superstore
WHERE profit<0;
### Output


-- Date Filter--

SELECT *
FROM superstore

### Output




## Step 4: Aggregation

-- Category-wise Sales--

SELECT category,
SUM(sales) AS Total_Sales
FROM superstore
GROUP BY category;
### Output


-- Region-wise Average Sales--

SELECT region,
AVG(sales) AS Average_Sales
FROM superstore
GROUP BY region;
### Output


-- Segment-wise Order Count--

SELECT segment,
COUNT(*) AS Total_Orders
FROM superstore
GROUP BY segment;
### Output


-- Category-wise Quantity Sold--

SELECT category,
SUM(quantity) AS Total_Quantity
FROM superstore
GROUP BY category;
### Output


## Step 5: Sort and limit results (top products, top categories)

-- Top 10 Products by Sales

SELECT product_name,
SUM(sales) AS Total_Sales
FROM superstore
GROUP BY product_name
ORDER BY Total_Sales DESC
LIMIT 10;
### Output


-- Top 10 Customers by Sales

SELECT customer_name,
SUM(sales) AS Total_Sales
FROM superstore
GROUP BY customer_name
ORDER BY Total_Sales DESC
LIMIT 10;
### Output


-- Top 5 Cities by Profit

SELECT city,
SUM(profit) AS Total_Profit
FROM superstore
GROUP BY city
ORDER BY Total_Profit DESC
LIMIT 5;
### Output


### Step 6: Solve use cases (monthly trends, top customers, duplicates).
-- Monthly Sales Trend

SELECT MONTH(order_date) AS Month,
SUM(sales) AS Total_Sales
FROM superstore
GROUP BY MONTH(order_date)
ORDER BY Month;
### Output


-- Top Customers

SELECT customer_name,
       SUM(sales) AS Total_Sales
FROM superstore
GROUP BY customer_name
ORDER BY Total_Sales DESC
LIMIT 10;
### Output


-- Duplicate Orders

SELECT order_id,
COUNT(*) AS Duplicate_Count
FROM superstore
GROUP BY order_id
HAVING COUNT(*) > 1;
### Output


-- Highest Profit Products

SELECT product_name,
SUM(profit) AS Total_Profit
FROM superstore
GROUP BY product_name
ORDER BY Total_Profit DESC
LIMIT 10;



## Step 7:Validate results (row counts, data quality)

-- Total Rows--

SELECT COUNT(*) AS Total_Rows
FROM superstore;
### Output


-- Check NULL Values--

SELECT COUNT(*) AS Null_Sales
FROM superstore
WHERE sales IS NULL;
### Output


SELECT COUNT(*) AS Null_Profit
FROM superstore
WHERE profit IS NULL;
### Output


-- Check Duplicate Order IDs--

SELECT order_id,
COUNT(*) AS Duplicate_Count
FROM superstore
GROUP BY order_id
HAVING COUNT(*) > 1;
### Output
